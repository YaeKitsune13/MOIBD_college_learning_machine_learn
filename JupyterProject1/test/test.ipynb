```python
NOC = pd.DataFrame(df2['NOC'].value_counts().head(40))
plt.figure(figsize = (14,20))
sb.barplot(data=NOC, x = 'count', y = 'NOC', palette='viridis', hue='count');
```

# Посмотрим какие страны какие игры предпочитают


```python
top_countries = df2['NOC'].value_counts().head(20).index
top_sports = df2['Sport'].value_counts().head(15).index

ct = pd.crosstab(df2['NOC'], df2['Sport'])
ct_top = ct.loc[top_countries, top_sports]

plt.figure(figsize=(14,10));
ct_top.plot(kind='barh', stacked=True, figsize=(14,10), width=0.8,cmap='viridis');
plt.title('Распределение видов спорта по топ-20 стран')
plt.xlabel('Количество')
plt.legend(bbox_to_anchor=(1.05,1), loc='upper left')
```

> лидеры (USA, FRA, ITA, GBR, CAN) представлены в большом числе видов спорта.

> атлетика — самая массовая дисциплина в большинстве стран.
